```python
import pandas as pd
import os
import numpy as np
data = pd.read_csv('weather.csv')
data
```

```python
data.info()
```

```python
# Преобразолвание даты в нормальный вид

data['DATE'] = pd.to_datetime(data['DATE'], format='%Y%m%d')
```

```python
#Заменим Nan в атрибуте MONTH на месяц, расположенный в атрибуте DATE

for i in range(0,3654):
    m=data.loc[i,'DATE'].month
    t = data.loc[i,'MONTH']
    is_nan=np.float64(t)
    if np.isnan(is_nan):
        data.loc[i,'MONTH'] = m
data
```

```python
#Заменим тип атрибута MONTH на int
data['MONTH'] = data['MONTH'].astype('int64')

data.info()
```

```python
#Просмотрев все атрибуты были обнаружены выбросы

#Выбросы в атрибудет BASEL_temp_mean
data['BASEL_temp_mean'] = data['BASEL_temp_mean'].astype('str')
print(data['BASEL_temp_mean'].unique())

data['BASEL_temp_mean'] = data['BASEL_temp_mean'].replace('219.0', '21.9')
data['BASEL_temp_mean'] = data['BASEL_temp_mean'].replace('-93.0', '-9.3')
data['BASEL_temp_mean'] = data['BASEL_temp_mean'].replace('123.0', '12.3')
```

```python

print(data['BASEL_temp_mean'].unique())
data['BASEL_temp_mean'] = data['BASEL_temp_mean'].astype('float64')
```

```python
#Выбросы в атрибудет BASEL_humidity
data['BASEL_humidity'] = data['BASEL_humidity'].astype('str')
print(data['BASEL_humidity'].unique())
```

```python
data['BASEL_humidity'] = data['BASEL_humidity'].replace('8.1', '0.81')
data['BASEL_humidity'] = data['BASEL_humidity'].replace('45.0', '0.45')
data['BASEL_humidity'] = data['BASEL_humidity'].replace('9.5', '0.95')
data['BASEL_humidity'] = data['BASEL_humidity'].replace('49.0', '0.49')
print(data['BASEL_humidity'].unique())
```

```python
#Заменим все Nan в атрибуте BASEL_temp_mean

for i in range(0,3654):
    mean_max = data.loc[i,'BASEL_temp_max']
    mean_min = data.loc[i,'BASEL_temp_min']
    t=data.loc[i,'BASEL_temp_mean']
    is_nan=np.float64(t)
    if np.isnan(is_nan):
        data.loc[i,'BASEL_temp_mean'] = (mean_max+mean_min)/2
data

```
